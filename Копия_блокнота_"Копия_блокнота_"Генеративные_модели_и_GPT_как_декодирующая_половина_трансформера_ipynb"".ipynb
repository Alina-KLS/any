```python
import numpy as np
import random
```

```python
random.seed(1234)
np.random.seed(1234)
```

```python
val_ind = random.sample(range(data.shape[0]), 5000)
```

```python
train = [data.iloc[i]['review_text'] for i in range(len(data)) if i not in val_ind]
valid = [data.iloc[i]['review_text'] for i in range(len(data)) if i in val_ind]
```

```python
len(train), len(valid)
```

```python
with open("train.txt", "w") as file:
    file.write("\n".join(train))
```

```python
with open("valid.txt", "w") as file:
    file.write("\n".join(valid))
```
